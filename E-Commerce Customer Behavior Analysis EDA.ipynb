# Conclusion

The bivariate analysis reveals several factors that are associated with customer satisfaction.

- Customers from **Chicago** and **Miami** exhibit the highest proportion of dissatisfied responses.
- **Bronze Membership** customers consistently report lower satisfaction than Gold members.
- Customers who **used discount coupons** tend to report lower satisfaction than those who did not.
- Customers aged **35 years and above** also appear to report lower satisfaction more frequently.
- Satisfied customers generally spend more, purchase more items, provide higher ratings, and return to the platform sooner than dissatisfied customers.

Overall, customer satisfaction appears to be associated with multiple customer characteristics rather than a single factor. These findings provide a strong foundation for the upcoming multivariate analysis, where the combined effect of multiple variables on customer satisfaction can be explored in greater detail.

## Multivariate Analysis

### 1. Does spending behavior differ across membership types and satisfaction levels?

- Overall, **satisfied customers tend to spend more**, particularly among **Gold** and **Silver** membership holders.
- Within the **Bronze** membership, however, **unsatisfied customers appear to spend slightly more than neutral customers**.
- This suggests that customer satisfaction is generally associated with higher spending, although this pattern varies across membership tiers.

---

### 2. Do satisfied customers purchase more items for the same spending level?

- Yes, the visualization clearly indicates that **satisfied customers generally purchase more items** than neutral or unsatisfied customers at similar spending levels.
- This suggests a positive relationship between customer satisfaction, purchase quantity, and overall spending behavior.

---

### 3. Does age influence spending differently for satisfied and dissatisfied customers?

- Yes, the relationship varies across different age groups.
- Customers in the **28–31 year** age range appear to have a higher concentration of satisfied responses while also exhibiting higher spending.
- In contrast, customers outside this age range show a greater mix of **Neutral** and **Unsatisfied** responses.
- Although a general pattern is visible, additional data would be required to determine whether age itself is the primary factor influencing customer spending and satisfaction.

---

# Conclusion

The multivariate analysis strengthens the findings from the bivariate analysis by showing that **customer satisfaction is associated with multiple factors simultaneously** rather than a single variable.

Key observations include:

- Satisfied customers generally spend more across most membership types.
- Customers with higher satisfaction tend to purchase more items.
- The **28–31 year** age group appears to be the most engaged, exhibiting both higher spending and higher satisfaction levels.

Overall, these results suggest that **customer satisfaction is closely linked with purchasing behavior, membership type, and age**, providing a more comprehensive understanding of customer behavior than individual or pairwise analyses alone.
